LINEAR REGRESSION

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
path="/content/drive/MyDrive/Data science notes _ myself/projects/diabetes.final3 - diabetes  - diabetes - Sheet1.csv.csv.csv"

raw_data = pd.read_csv(path)

x = raw_data[['No_Pation', 'Pregnancies', 'AGE', 'Glucose', 'SkinThickness','BloodPressure',  'Insulin', 'BMI', 'DiabetesPedigreeFunction']]
y = raw_data['Outcome']
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)

print(model.coef_)

print(model.intercept_)

pd.DataFrame(model.coef_, x.columns, columns = ['Coeff'])

predictions = model.predict(x_test)

# plt.scatter(y_test, predictions)

plt.hist(y_test - predictions)

from sklearn import metrics

metrics.mean_absolute_error(y_test, predictions)

metrics.mean_squared_error(y_test, predictions)

np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

DecisionTreeClassifier

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
path="/content/drive/MyDrive/Data science notes _ myself/projects/diabetes.final3 - diabetes  - diabetes - Sheet1.csv.csv.csv"

data_set = pd.read_csv(path)

x = data_set[['No_Pation', 'Pregnancies', 'AGE', 'Glucose', 'SkinThickness','BloodPressure',  'Insulin', 'BMI', 'DiabetesPedigreeFunction']]
y = data_set['Outcome']
#splitting dataset
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3 , random_state = 0)
# feature scaling
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
x_train = Sc.fit_transform(x_train)
x_test = Sc.fit_transform(x_test)

from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
from sklearn import metrics
pre = dtree.predict(x_test)
print("Accuracy Score =", format(metrics.accuracy_score(y_test,pre)))
```

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
path="/content/drive/MyDrive/Data science notes _ myself/projects/diabetes.final3 - diabetes  - diabetes - Sheet1.csv.csv.csv"

data_set = pd.read_csv(path)
x = data_set[['No_Pation', 'Pregnancies', 'AGE', 'Glucose', 'SkinThickness','BloodPressure',  'Insulin', 'BMI', 'DiabetesPedigreeFunction']]
y = data_set['Outcome']
#splitting dataset
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3 , random_state = 0)
# feature scaling
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
x_train = Sc.fit_transform(x_train)
x_test = Sc.fit_transform(x_test)
# creating classifier
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
# import scikit metrics
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test,y_pred)
# model accuracy
print("accu: ",metrics.accuracy_score(y_test,y_pred))
```
